# 1.2 Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
# Need to remove NA
# Regression - Try to predict PM2.5 (y) from non PM2.5 columns (X)
# parameter: data, n(rows to shift), dependent_col, drop_cols(including dependent_col)
def valid(col,drop_cols):
    return col not in drop_cols

def randomforest(data,n,dependent_col,drop_cols):
    X = data[[x for x in data.columns if valid(x,drop_cols)]]
    y = data[dependent_col]
    if n>0:
        X = X.iloc[:-n]
        y = y.shift(-n).dropna()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
    # Create a model 
    rf_model = RandomForestRegressor()
    # Train the model
    rf_model.fit(X_train, y_train)
    
    return rf_model,X_test, y_test
```

# 1.3 Use Random Forest for PM2.5 Concentration

```python
data_shunyi
```

```python
data_concentration = data_shunyi.dropna()
data_concentration[:10]
```

```python
#recall random forest
drop_cols = ['No','year','day','hour','PM2.5','PM10','station']
dependent_col = 'PM2.5'
rf_model,X_test, y_test = randomforest(data_concentration,1,dependent_col,drop_cols)
print("accurate:", rf_model.score(X_test, y_test))
print()
print("predicted data and real data:")
print(pd.concat([pd.Series(rf_model.predict(X_test)), y_test.reset_index(
drop=True)], axis=1))
print()
print("feature importances:")
print(sorted(zip(X_test.columns, rf_model.feature_importances_),
    key=lambda x: x[1], reverse=True))
```
